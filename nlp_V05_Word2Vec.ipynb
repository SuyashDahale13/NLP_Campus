```python
import gensim 
import os 
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess

story = []
for filename in os.listdir('data'):
    
    f = open(os.path.join('data',filename),encoding='unicode_escape')
    print(f)
    corpus = f.read()
    raw_sent = sent_tokenize(corpus)
    for sent in raw_sent:
        story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story
```

```python
model = gensim.models.Word2Vec(
    window=10, # both side will be 10-10 words of target word.
    min_count=2, # wahi sentences lenge jinme atleast two words ho.
    vector_size=100, # output vectors ka size jo chahiye ho wo.
)
```

```python
model.build_vocab(story) # unique words from our corpus.
```

```python
model.train(story, total_examples=model.corpus_count, epochs=model.epochs) # corupus, total sentences in corpus, no of epochs
```

```python
model.wv.most_similar('daenerys') # wv = word2vec
```
